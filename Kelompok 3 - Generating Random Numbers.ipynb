# A. TUGAS 1

1. Kode Script Program

```python
import random
import numpy as np
import matplotlib.pyplot as plt
#fungsi kejadian pelemparan koin, muncul head : 0 dan muncul tail : 1
def kejadian_koin():
    return random.randint(0,1)
kejadian_koin()
```

```python
#FUNGSI MONTE CARLO SIMULASI

#tempat untuk nilai probabilitas
tempat_daftar = []

def monte_carlo(jumlah):
    hasil_kejadian = 0
    for i in range(jumlah):
        hasil_pelemparan = kejadian_koin()
        hasil_kejadian = hasil_kejadian + hasil_pelemparan
        
        #hitung nilai probabilitasnya
        nilai_probabilitas = hasil_kejadian/(i+1)
        
        #tambahkan nilai dari probabilitas ke list yang sudah ada
        tempat_daftar.append(nilai_probabilitas)
        
        #membuat plot hasil dari kejadian yang sudah ada
        plt.axhline(y=0.5, color='r', linestyle='-')
        plt.xlabel("iterasi")
        plt.ylabel("Probabilitas")
        plt.plot(tempat_daftar)   
    return hasil_kejadian/jumlah
```

2. Output Program

a) 10 Iterasi

```python
prob_10 = monte_carlo(10)
print("Nilai probabilitas dari 10 iterasi:", prob_10)
```

b) 50 Iterasi

```python
prob_50 = monte_carlo(50)
print("Nilai probabilitas dari 50 iterasi:", prob_50)
```

c) 250 Iterasi

```python
prob_250 = monte_carlo(250)
print("Nilai probabilitas dari 250 iterasi:", prob_250)
```

d) 500 Iterasi

```python
prob_500 = monte_carlo(500)
print("Nilai probabilitas dari 500 iterasi:", prob_500)
```

e) 100 Iterasi

```python
prob_1000 = monte_carlo(1000)
print("Nilai probabilitas dari 1000 iterasi:", prob_1000)
```

1. Distribusi Uniform

```python
#penggunaan library
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

#objek random generator
rng = np.random.default_rng(100)

#membangkitkan 1000 bilangan acak dengan batas bawah 1 dan batas atas 11
rand_uniform = rng.uniform(low = 1, high = 11, size = 1000)

#di tampilkan hasil
print("Di Tampilkan hasil bilangan random distribusi uniform : \n", rand_uniform)
```

```python
plt.figure(figsize = (8, 4))

sns.set_style("ticks")

#histogram dari bilangan acak yang di bangkitkan
sns.histplot(data = rand_uniform, stat = "density", color = "darkcyan", alpha = 0.6)

#kurva sebaran teoritis dengan batas bawah 1 dan batas atas 11
#pada kisaran nilai x 1-10
x = np.linspace(1, 10, 1000)
y = stats.uniform.pdf(x, loc = 1, scale = 11)
sns.lineplot(x = x, y = y, color = "red")

plt.show()
```

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def pdf_beta(x):
    return 6 * x * (1 - x)

def pdf_proposal(x):
    return 1

def beta_per_prop(x):
    return pdf_beta(x) / (M * pdf_proposal(x))

M = 6
n_samples = 1000
samples = []

while len(samples) < n_samples:
    x = np.random.rand()
    u = np.random.rand()

    if u <= beta_per_prop(x):
        samples.append(x)
samples = np.array(samples)
print(samples)
```

```python
plt.figure(figsize=(8, 4))
sns.set_style("ticks")
sns.histplot(data=samples, stat="density", color="grey", alpha=0.6)
x = np.linspace(0, 1, 1000)
y = pdf_beta(x)
sns.lineplot(x=x, y=y, color="red")
plt.show()
```
